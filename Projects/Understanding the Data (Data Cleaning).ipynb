# Normalizing Data

```python
import pandas as pd
df = pd.read_csv("artwork_sample.csv")
df.head(3)
```

```python
df.dtypes
```

```python
df.height.min()
```

```python
df.height.std()
```

```python
height = df.height

norm = (height-height.mean())/height.std()
norm
```

```python
h = df.height

MinMax = (h-h.min())/(h.max()-h.min())
MinMax
```

```python
norm.min()
```

```python
norm.max()
```

```python
MinMax.min()
```

```python
MinMax.max()
```

```python
df.height = MinMax
```

```python
df["std_height"] = norm
df.head()
```
